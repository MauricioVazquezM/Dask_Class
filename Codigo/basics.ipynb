## Introduccion a las bases de Dask

```python
# Funciones basicas
import dask
from time import sleep

def inc(x):
    sleep(1)
    return x + 1

def add(x, y):
    sleep(1)
    return x + y
```

```python
%%time

x = inc(2)
y = inc (3)
z = add(x,y)
```

```python
from dask import delayed

@delayed
def inc(x):
    sleep(1)
    return x + 1

@delayed
def add(x, y):
    sleep(1)
    return x + y
```

```python
%%time

x = delayed(inc)(2)
y = delayed(inc)(3)
z = delayed(add)(x,y)
```
